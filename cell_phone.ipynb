```python
import numpy as np
import pandas as pd
import config1

from sklearn.model_selection  import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.metrics import  mean_absolute_error ,mean_squared_error,r2_score

import warnings
warnings.filterwarnings('ignore')

import json
import pickle
```

```python
dir(config1)
```

```python
xlsx_file_path = config1.XLSX_FILE_PATH
df = pd.read_excel(xlsx_file_path)
df
```

```python
df['cpu core'].value_counts()
```

```python
df1 =df.drop('Product_id',axis = 1)
df1
```

```python
df1.rename(columns = {'cpu core':'cpu_core','cpu freq':'cpu_freq','internal mem':'internal_mem'},inplace=True)
df1.info()
```

```python
x = df1.drop('Price',axis = 1)
y=  df1['Price']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=15)
x_train
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
```

```python
y_pred_train = lin_reg.predict(x_train)

mse = mean_squared_error(y_train, y_pred_train)
print("Mean Squared Error :",mse)

rmse = np.sqrt(mse)
print("Root Mean Squared Error :",rmse)

mae = mean_absolute_error(y_train, y_pred_train)
print("Mean Absolute Error :",mae)

r_squared = r2_score(y_train, y_pred_train)
print("R Squared Value is :",r_squared)
```

```python
y_pred_test = lin_reg.predict(x_test)
mse = mean_squared_error(y_test, y_pred_test)
print("Mean Squared Error :",mse)

rmse = np.sqrt(mse)
print("Root Mean Squared Error :",rmse)

mae = mean_absolute_error(y_test, y_pred_test)
print("Mean Absolute Error :",mae)

r_squared = r2_score(y_test, y_pred_test)
print("R Squared Value is :",r_squared)
```
